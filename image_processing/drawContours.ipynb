```python
image_path = "img/igo.jpg"
img = cv2.imread(image_path)
```

```python
show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
```

```python
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(imgray, 'gray')
```

```python
ret,thresh = cv2.threshold(imgray,127,255,0)
plt.imshow(thresh)
```

```python
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
plt.imshow(image)
```

```python
rect_and_contours = np.copy(img)
min_rect_area = 10000
_,rect_contours,_= cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
large_contours = [cnt for cnt in rect_contours if cv2.contourArea(cnt) > min_rect_area]
drawed_image = cv2.drawContours(rect_and_contours, large_contours, -1, (255,0,0))
```

```python
plt.imshow(drawed_image, "gray")
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv, "gray")
```

```python
lower = 110,50,50
upper = 130,255,255
img = cv2.inRange(hsv, lower, upper)
plt.imshow(img)
```
