```
l = ["I ", True,"Python"]

try:
  "".join(l)
except:
  print("error!!")
  
```

```
#try入れる
import sys
l = ["I ", True,"Python"]


try:
  "".join(l)
except:
  print(sys.exc_info())
  print("====")
  print("Unexpected error:", sys.exc_info()[0])
  
```

```
#リストでない場合のassert入れる
import sys
l = "I "

try:
  assert isinstance(l, list), "this is not list object"
  "".join(l)
except AssertionError as error:
     print ("AssertionErrorが発生しました。", error)
except:
  print(sys.exc_info())
  print("====")
  print("Unexpected error:", sys.exc_info()[0])
```

```
#リスト内のチェックassert入れる
import sys
l = ["I ", True,"Python"]

try:
  assert isinstance(l, list), "this is not list type"
  assert all(isinstance(x, str) for x in l), "this element is not string type"
  "".join(l)
except AssertionError as error:
     print ("AssertionErrorが発生しました。", error)
except:
  print(sys.exc_info())
  print("====")
  print("Unexpected error:", sys.exc_info()[0])
```

```
#通常に動かす
import sys
l = ["I ", "True ","Python"]

try:
  assert isinstance(l, list), "this is not list type"
  assert all(isinstance(x, str) for x in l), "this element is not string type"
  print("".join(l))
except AssertionError as error:
     print ("AssertionErrorが発生しました。", error)
except:
  print(sys.exc_info())
  print("====")
  print("Unexpected error:", sys.exc_info()[0])
```
