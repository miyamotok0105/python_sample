```
#try catch 例外処理
```

```
l = ["I ",True,"Python"]
try:
  "".join(l)
except:
  print("error!")
  #文字列じゃないのがきた時の処理を書いておく
```

```
import sys
l = ["I ",True,"Python"]
try:
  "".join(l)
except:
  print("error!")
  print(sys.exc_info())
```

```
import sys
# l = ["I ",True,"Python"]
l = "  "
try:
  assert isinstance(l , list), " this object is not list"
  "".join(l)
except:
  print("error!")
  print(sys.exc_info())
```

```
import sys
# l = ["I ",True,"Python"]
l = "  "
try:
  assert isinstance(l , list), " this object is not list"
  "".join(l)
except AssertionError as error:
  print("Assertion error!")
except:
  print("error!")
  print(sys.exc_info())
```
